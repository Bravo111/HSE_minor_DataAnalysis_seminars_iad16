### Знакомство с IPython Notebook

IPython notebook состоит из ячеек (cells), в каждой из которых может содержать либо код программы (code), либо текст (markdown).

Текст, который вы сейчас читаете, тоже находится в ячейке. Чтобы получить возможность редктировать его, достаточно кликнуть два раза по ячейке (либо выделить её и нажать **enter**), затем нажать **shift + enter**. 

Правила вёрстки текста можно найти по [по ссылке](https://help.github.com/articles/markdown-basics/).

Ячейки с кодом:

```python
def my_sum(a, b):
    return a + b

x = my_sum(2, 3)
print(x)
```

У ячеек одинаковая область видимости, то есть в новой ячейке можно использовать значения, вычисленные в других ячейках:

```python
print(2 * x)
```

Чтобы создать новую ячейку под выделенной, нужно либо нажать на **+** в левом верхнем углу меню, либо воспользоваться горячей клавишей **b**.

### Знакомство с matplotlib

Эта строка нужна, чтобы графики строились внутри ipython notebook`а:

```python
%pylab inline
```

```python
import matplotlib.pylab as plt
import math
import random
```

На следующих семинарах для хранения данных мы будем использовать массивы из библиотеки *NumPy*, но сейчас обойдёмся стандартными списками языка *Python*:

```python
x_arr = [x / 100.0 for x in range(-1000, 1000)]
sinx_arr = [math.sin(x) for x in x_arr]
cosx_arr = [math.cos(x) for x in x_arr]
tanx_arr = [math.tan(x) for x in x_arr]
print(x_arr[:10])
```

Построим простой график функции $\sin x$:

```python
plt.plot(x_arr, sinx_arr)
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('y')
```

Теперь добавим ещё и функцию $\cos x$. 

Чтобы различать графики, с помощью функции plt.legend добавим легенду графика.

```python
plt.plot(x_arr, sinx_arr)
plt.plot(x_arr, cosx_arr)
plt.title('sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['sin(x)', 'cos(x)'], loc='upper right')
```

Как можно было заметить, у функции plt.legend есть параметр loc, который отвечает за расположение легенды. 

Этот параметр может принимать ограниченное количество значений, помнить которые не нужно: но важно уметь быстро находить их в документации.

Графики, которые мы получили, не очень похожи на привычные нам. Дело в том, что оси координат имеют разный масштаб. 

Во-первых, исправим это с помощью вызова функции plt.axis('equal'). 

Во-вторых, изменим размер графика.

```python
plt.figure(figsize=(15,4))
plt.plot(x_arr, sinx_arr)
plt.plot(x_arr, cosx_arr)
plt.title('sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.legend(['sin(x)', 'cos(x)'])
```

Заметим, что на текущий момент подписи к графикам слишком маленького размера.

Исправить эту проблему поможет, например, вот такой кусок кода:

```python
from matplotlib import rc

font = {'size': 13}
rc('font', **font)
```

##### Итого:

```python
plt.figure(figsize=(15,3))
plt.plot(x_arr, sinx_arr)
plt.plot(x_arr, cosx_arr)
plt.title('sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.legend(['sin(x)', 'cos(x)'])
plt.savefig('./sincos.png', frameon=True)
```

##### Несколько графиков на одной картинке

```python
plt.figure(figsize=(10, 10))
m = 3
n = 3
for idx in range(m * n):
    plt.subplot(m, n, idx + 1)
    plt.title('Plot #' + str(idx + 1))
    plt.plot(x_arr, sinx_arr, 'b')
```
